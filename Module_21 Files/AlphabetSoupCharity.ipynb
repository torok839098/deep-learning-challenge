## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from keras.callbacks import ModelCheckpoint

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(columns=['EIN', 'NAME'])
```

```python
#From statology
for col in application_df:
  print(application_df[col].unique())
```

```python
# Other method to  Determine the number of unique values in each column.

application_df.nunique()
```

```python
#Look at APPLICATION_TYPE value counts for binning
Application_TypeCounts=application_df.value_counts('APPLICATION_TYPE')
print(Application_TypeCounts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace=list (Application_TypeCounts[Application_TypeCounts < 500].index)
print(application_types_to_replace)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# Replace in dataframe
# Check to make sure binning was successful

application_types_to_replace=list (Application_TypeCounts[Application_TypeCounts < 500].index)
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
Classification_ValueCounts=application_df['CLASSIFICATION'].value_counts()
print(Classification_ValueCounts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
Classification_countsRevised=Classification_ValueCounts[Classification_ValueCounts>1]
print(Classification_countsRevised)
```

```python
#classifications_to_replace=list(Classification_countsRevised[Classification_countsRevised < 800].index)
#print(classifications_to_replace)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace

classifications_to_replace=list(Classification_ValueCounts[Classification_ValueCounts < 1000].index)
# Replace in dataframe
#for cls in classifications_to_replace
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Split our preprocessed data into our features and target arrays

y = application_df['IS_SUCCESSFUL']
X = application_df.drop(columns= ['IS_SUCCESSFUL','EIN','NAME'])

# Preview the data
X[:5]
```

```python
y[:5]
```

```python
# Convert categorical data to numeric with `pd.get_dummies
X = pd.get_dummies(X)
```

```python
#Display features
X.head()
```

```python

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python

# Get trained shape
X_train.shape
```

```python

# Get test shape
X_test.shape
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

#nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE

# Second hidden layer
#  YOUR CODE GOES HERE

# Output layer
#  YOUR CODE GOES HERE

# Check the structure of the model
#nn.summary()
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()
```

```python
# First hidden layer
nn.add(tf.keras.layers.Dense(input_dim=43, units=80, activation="relu"))
# Second hidden layer
nn.add(tf.keras.layers.Dense(input_dim=10, units=30, activation="relu"))
# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="linear"))
nn.summary()
```

```python
#Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
#Create a callback that saves the model's weights every five epochs
checkpoint = ModelCheckpoint("best_model.hdf5", monitor='loss', verbose=1,save_best_only=True, mode='auto', period=5)
fit_model=nn.fit(X_train_scaled, y_train,
          epochs=100,
          verbose=1,
          validation_data=(X_test, y_test),
          callbacks=[checkpoint])

#fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
